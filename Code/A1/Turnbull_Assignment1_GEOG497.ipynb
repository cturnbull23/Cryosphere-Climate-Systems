```python
# Import Modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read in the csv file and create a dataframe
Zachos_df = pd.read_fwf('zachos2001.txt', skiprows=list(range(0,88)), index_col='Age (Ma)')
#Zachos_df
```

```python
# Create a figure and plot for Zachos Dataset
fig, ax = plt.subplots(figsize=(10,10))

# .index.values gives the values of the index_col, which = 'Age (Ma)'
Zachos = plt.plot(Zachos_df.index.values, Zachos_df['d18O adj.*'])

# Reverse order of values
ax.set_xlim(max(Zachos_df.index.values), min(Zachos_df.index.values))

# Set labels and grid
ax.set_xlabel('Millions of Years Ago')
ax.set_ylabel('Benthic Delta-O-18 (per mill)')
ax.grid()

# Set title and save figure
plt.title('Delta-O-18 Evolution (Zachos Dataset)')
plt.savefig("figures/ZachosDeltaO18.png", facecolor='white', dpi=300)
```

```python
# Read in the csv file and create a dataframe
#DomeFuji_df = pd.read_fwf('DomeFuji.txt', index_col='Time')
#DomeFuji_df

DomeFuji_df = pd.read_csv('DomeFuji.csv', index_col='Age')
#DomeFuji_df
```

```python
# Create a figure and plot for Dome Fuji Dataset
fig, ax = plt.subplots(figsize=(10,10))


DomeFuji = plt.plot(DomeFuji_df.index.values, DomeFuji_df['del_18O'])
ax.set_xlim(max(DomeFuji_df.index.values), min(DomeFuji_df.index.values))


ax.set_xlabel('Years Ago')
ax.set_ylabel('Benthic Delta-O-18 (per mill)')
ax.grid()

plt.title('Delta-O-18 Evolution (Dome Fuji Dataset)')
plt.savefig("figures/DomeFujiDeltaO18.png", facecolor='white', dpi=300)
```

```python
Renland_df = pd.read_csv('Renland.csv', index_col='Age (yrs b2k)')
#Renland_df
```

```python
# Create a figure and plot for Dome Fuji and Renland Datasets
fig, ax = plt.subplots(figsize=(10,10))


DomeFuji = plt.plot(DomeFuji_df.index.values, DomeFuji_df['del_18O'], label='Dome Fuji')
Renland = plt.plot(Renland_df.index.values, Renland_df['Renland d18O'], label='Renland')
ax.set_xlim(max(Renland_df.index.values), min(DomeFuji_df.index.values))


ax.set_xlabel('Thousands of Years Ago')
ax.set_ylabel('Benthic Delta-O-18 (per mill)')
ax.grid()

plt.title('Delta-O-18 Evolution (Dome Fuji and Renland Datasets)')
plt.legend()
plt.savefig("figures/DomeFuji-Renland-DeltaO18.png", facecolor='white', dpi=300)
```

```python
RenlandStack_df = pd.read_csv('RenlandStack.csv', index_col='Age')
#RenlandStack_df
```

```python
# Create a figure and plot for the Renland Stack Dataset
fig, ax = plt.subplots(figsize=(10,10))

# Plot both summer and winter d18O values
RenlandSummer = plt.plot(RenlandStack_df.index.values, RenlandStack_df['d18O Summer'], label='Summer')
RenlandWinter = plt.plot(RenlandStack_df.index.values, RenlandStack_df['d18O winter'], label='Winter')

# Create and plot a 10-year rolling mean for both summer and winter
RenlandStack_df['rolling_d18O_summer'] = RenlandStack_df['d18O Summer'].rolling(10).mean()
RenlandRollingSummer = plt.plot(RenlandStack_df.index.values, RenlandStack_df['rolling_d18O_summer'], 
                          label='10-Year Rolling Mean Summer', linewidth=2, color='magenta')

RenlandStack_df['rolling_d18O_winter'] = RenlandStack_df['d18O winter'].rolling(10).mean()
RenlandRollingWinter = plt.plot(RenlandStack_df.index.values, RenlandStack_df['rolling_d18O_winter'], 
                          label='10-Year Rolling Mean Winter', linewidth=2, color='Green')



ax.set_xlabel('Year (1801 AD - 2014 AD)')
ax.set_ylabel('Benthic Delta-O-18 (per mill)')
ax.grid()

plt.title('Delta-O-18 Evolution (Renland Stack Dataset - Summer/Winter)')
plt.legend()
plt.savefig("figures/RenlandStack-DeltaO18.png", facecolor='white', dpi=300)
```

```python
LR04_df = pd.read_csv('LR04.csv', index_col='Time')
DomeFuji_df = pd.read_csv('DomeFuji.csv', index_col='Age')


# Create a figure and plot for LR04 and Dome Fuji Datasets
fig, ax = plt.subplots(figsize=(10,10))

# Plot LR04, set x-limits, and create axes legend
LR04 = plt.plot(LR04_df.index.values, LR04_df['d18O'], label='LR04')
ax.set_xlim(max(DomeFuji_df.index.values), min(DomeFuji_df.index.values))

ax.legend(loc='upper left')
ax.set_xlabel('Years Ago')
ax.set_ylabel('Benthic Delta-O-18 (per mill)')
ax.grid()

# Create second y-axis to plot Dome Fuji Dataset
ax2 = ax.twinx()

DomeFuji = ax2.plot(DomeFuji_df.index.values, DomeFuji_df['del_18O'], 
                    color='orange', label='Dome Fuji')

ax2.legend(loc='upper right')
ax2.set_ylabel('Benthic Delta-O-18 (per mill)', rotation=270, labelpad=12)


plt.title('Delta-O-18 Evolution (LR04 and Dome Fuji Datasets)')
plt.savefig("figures/LS04_DomeFuji_DeltaO18.png", facecolor='white', dpi=300)
```
